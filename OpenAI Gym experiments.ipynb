```python
import gym
import time
import matplotlib.pyplot as plt
```

## [CartPole experiment](https://gym.openai.com/envs/CartPole-v0/)

> A pole is attached by an un-actuated joint to a cart, which moves along a frictionless track. The system is controlled by applying a force of +1 or -1 to the cart. The pendulum starts upright, and the goal is to prevent it from falling over. A reward of +1 is provided for every timestep that the pole remains upright. The episode ends when the pole is more than 15 degrees from vertical, or the cart moves more than 2.4 units from the center.


Remember; A state s is a complete description of the state of the world. There is no information about the world which is hidden from the state. An observation o is a partial description of a state, which may omit information.

```python
env = gym.make('CartPole-v1')
num_episodes = 50
steps_total = []

for i in range(num_episodes):
    observation = env.reset()
    steps = 0
    
    while True:
        steps += 1
        
        # You can either go left or right: {0, 1}
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)

        # Show related information,
        # print(observation, reward, done, info)
        # If you want to see the full episode,
        # env.render()
        
        if done:
            steps_total.append(steps)
            print("Episode finished after %i steps" % steps)
            break

print("Average number of steps %.2f" % (sum(steps_total)/num_episodes))
plt.plot(steps_total)
plt.show()
```

## [FrozenLake experiment](https://gym.openai.com/envs/FrozenLake-v0/)

> The agent controls the movement of a character in a grid world. Some tiles of the grid are walkable, and others lead to the agent falling into the water. Additionally, the movement direction of the agent is uncertain and only partially depends on the chosen direction. The agent is rewarded for finding a walkable path to a goal tile.

```python
env = gym.make('FrozenLake-v0')
num_episodes = 1
steps_total = []

print("Number of actions %i" % env.action_space.n)

for i in range(num_episodes):
    observation = env.reset()
    steps = 0
    
    while True:
        steps += 1

        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)

        # Show related information,
        # print(observation, reward, done, info)
        
        # We can see uncertainty is introduced,
        # This is called an sochastic environment
        print(info)
        # Slippery has to be turned off, if
        # to make it a deterministic environment
        
        time.sleep(0.4)
        
        # If you want to see the full episode,
        env.render()
        
        if done:
            steps_total.append(steps)
            print("Episode finished after %i steps" % steps)
            break
            
print("Average number of steps: %.2f" % (sum(steps_total)/num_episodes))
plt.plot(steps_total)
```
